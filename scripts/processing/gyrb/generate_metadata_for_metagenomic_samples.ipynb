This scripts generates a metadata sheet with 9000 plus individual metagenomic samples by reformatting the metadata from Passoli et al to be compatible with this study.

```python
import pandas as pd
import numpy as np
import os 
```

```python
root_dir = '/Volumes/AHN/captive_ape_microbiome/'
os.chdir(root_dir)

analysis_dir = 'results/gyrb/processing/metagenomic_samples/'
os.chdir(analysis_dir)
```

```python
metagenomic_metadata_file = 'metagenomic_metadata_all_samples.txt'
metagenomic_metadata = pd.read_csv(metagenomic_metadata_file,sep='\t')
metagenomic_metadata.head()
```

```python
metagenomic_metadata['X.SampleID'] = metagenomic_metadata.apply(lambda row: row['Study'] +'__' + row['Sample ID'],axis=1)
metagenomic_metadata['X.SampleID'] = metagenomic_metadata['X.SampleID'].replace('.','').replace('-','')
metagenomic_metadata['Old_SampleID'] = metagenomic_metadata['Sample ID']
metagenomic_metadata['dataset'] = metagenomic_metadata['Study']
metagenomic_metadata['genus_sp'] = 'Homo_sapiens'
metagenomic_metadata['subspecies']  = 'Homo_sapiens'
metagenomic_metadata['common_name'] = 'human'
metagenomic_metadata['uniqueid'] = 'NA'
metagenomic_metadata['site'] = metagenomic_metadata['Country']
metagenomic_metadata['site_code'] = metagenomic_metadata['Country']
metagenomic_metadata['country'] = metagenomic_metadata['Country']
metagenomic_metadata['captivity_status'] = np.where(metagenomic_metadata['Westernized']=='Yes', 'industrialized_human', 'non_industrialized_human') 
metagenomic_metadata['amplicon'] = 'META'
metagenomic_metadata['sample'] = metagenomic_metadata['Sample ID']
metagenomic_metadata['Description'] = metagenomic_metadata['captivity_status']
metagenomic_metadata_select = metagenomic_metadata[['X.SampleID','Old_SampleID','dataset',
                     'genus_sp','subspecies','common_name','uniqueid','site','site_code','country',
                     'captivity_status','amplicon','sample','Description']]

metagenomic_metadata_select.to_csv('metagenomic_metadata_all_samples_formatted.txt',sep='\t',index=None,header=None)
metagenomic_metadata_select.head()
```

```python
os.chdir(root_dir)
metagenomic_metadata = analysis_dir + 'metagenomic_metadata_all_samples_formatted.txt'
metagenomic_metadata
amplicon_metadata = 'metadata/metadata_gyrb_amp.txt'
amplicon_metagenomic_metadata = 'metadata/metadata_gyrb_amp_meta.txt'
os.system("AWK '{print $0}' "+amplicon_metadata+" "+metagenomic_metadata+" > "+amplicon_metagenomic_metadata)
```
