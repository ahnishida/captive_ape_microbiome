```python
import os
#import pandas as pd
from Bio import SeqIO
from Bio import Seq

os.getcwd()
os.chdir('/Volumes/AHN/captive_ape_microbiome')
```

```python
metadata = pd.read_csv('metadata/metadata_Bt_samples_gyrb.txt',sep='\t',index_col=None)
tax_table = pd.read_csv('results/gyrb_all/Bt/assigned_taxonomy/ASVs_filtered_taxonomy.txt',sep='\t',index_col=0)
asv_table = pd.read_csv('results/gyrb_all/Bt/ASVs_filtered_counts.tsv',sep='\t',index_col=0)
fasta = 'results/gyrb_all/Bt/phylogeny/ASVs_filtered_ref_fulllength.fasta.aln'
groups = ['non_western_human', 'western_human', 'wild_gorilla', 'wild_bonobo', 'wild_chimp']

def get_description(sample_name):
    description = metadata[metadata.index==sample_name]['Description'].to_string(index=False).replace(' ','')
    return(description)
def get_HR_clade(descriptions,numSamples):
    if len(descriptions) == 0:
        return('NA')
    elif len(descriptions) == 1:
        return(descriptions[0])
    else:
        return('notHR')

descDict = {}
for index,row in asv_table.iterrows():
    print(index)
    samples = list(asv_table.columns[row>0])
    numSamples = len(samples)
    descriptions = list(set([get_description(x) for x in samples]))
    print(descriptions)
    descriptions_HR = [x for x in descriptions if x in groups]
    HR_clade = get_HR_clade(descriptions_HR,numSamples)
    print(HR_clade)
    descDict[index] = HR_clade
```

```python
with open('results/gyrb_all/Bt/phylogeny/ASVs_filtered_ref_fulllength.fasta.aln') as original: 
    with open('results/gyrb_all/Bt/phylogeny/ASVs_filtered_ref_fulllength.fasta.aln.renamed', 'w') as renamed:
        records = SeqIO.parse(original, 'fasta')
        for record in records:
            if 'ASV' in record.id:  
                try:
                    print(record.id)
                    HR_clade = descDict[record.id]
                    record.id = record.id + '__' + HR_clade
                except:
                    record.id = record.id + '__' + 'notfound'
            SeqIO.write(record, renamed, 'fasta')
```

```python
from Bio import Phylo
tree = Phylo.read("results/gyrb_all/Bt/phylogeny/ASVs_filtered_ref_fulllength.fasta.aln.renamed.tre", "newick")

def all_parents(tree):
    parents = {}
    for clade in tree.find_clades(order='level'):
        for child in clade:
            parents[child] = clade
    return parents

parents = all_parents(tree)
```
