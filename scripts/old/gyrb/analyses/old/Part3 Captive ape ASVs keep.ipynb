```python
import os
import pandas as pd
from Bio import SeqIO
from Bio import Seq


os.getcwd()
os.chdir('/Volumes/AHN/captive_ape_microbiome')
```

```python
metadata = pd.read_csv('metadata/metadata_Bt_samples_gyrb.txt',sep='\t',index_col=None)
tax_table = pd.read_csv('results/gyrb_all/Bt/assigned_taxonomy/ASVs_filtered_taxonomy.txt',sep='\t',index_col=0)
asv_table = pd.read_csv('results/gyrb_all/Bt/ASVs_filtered_counts.tsv',sep='\t',index_col=0)

def get_sample_description(sample_name):
    description = metadata[metadata.index==sample_name]['Description'].to_string(index=False).replace(' ','')
    return(description)

def get_ASV_taxonomy(ASV):
    genus = tax_table[tax_table.index==ASV][['Genus']].to_string(index=False,header=False)
    family = tax_table[tax_table.index==ASV][['Family']].to_string(index=False,header=False)
    
    return(family+genus)
print(get_ASV_taxonomy("ASV_1"))

ASV_sampleName_dict = {}  
ASV_sampleType_dict = {} 
ASV_taxonomy_dict = {}

res = []
for index,row in asv_table.iterrows():
    sampleNames = list(asv_table.columns[row>0])
    sampleNum = len(sampleNames)
    captivesampleNum = len([x for x in sampleNames if 'cp.' in x])
    wildsampleNum = len([x for x in sampleNames if 'wd.' in x])
    humansampleNum = len([x for x in sampleNames if 'wd.' not in x and 'cp.' not in x])
    ASV_sampleName_dict[index] = sampleNames
    sampleDescriptions = list(set([get_sample_description(x) for x in sampleNames]))
    ASV_sampleType_dict[index] = sampleDescriptions
    ASV_taxonomy = get_ASV_taxonomy(index)
    ASV_taxonomy_dict[index] = ASV_taxonomy
    ASV_sites = list(set([x.split('.')[2] for x in sampleNames if 'cp.' in x]))
    numSites = len(ASV_sites)
    numSpecies = len(set([x.split('_')[-1] for x in sampleDescriptions if 'captive' in x]))
    print(index)
    print(sampleNum)
    print(captivesampleNum)
    print(wildsampleNum)
    print(humansampleNum)
    print(sampleNames)
    print(ASV_taxonomy)
    print(sampleDescriptions)
    print(ASV_sites)
    row = [index,sampleNum,captivesampleNum,wildsampleNum,humansampleNum,sampleNames,ASV_taxonomy,sampleDescriptions,ASV_sites,numSites,numSpecies]
    res.append(row)
    
```

```python
allASVs = pd.DataFrame(res, columns = ['ASV',
                                 'sampleNum',
                                 'captivesampleNum',
                                 'wildsampleNum',
                                 'humansampleNum',
                                 'sampleNames',
                                 'ASV_taxonomy',
                                 'sampleDescriptions',
                                 'captive_sites',
                                 'numSites',
                                 'numSpecies'
                                 ]
                                )  
captiveASVs = []
wildASVs = []
nwhASVs = []
whASVs = []
for index, row in allASVs.iterrows():
    for desc in row['sampleDescriptions']:
        if 'captive' in desc and row['ASV'] not in captiveASVs:
            captiveASVs.append(row['ASV'])
        if 'wild' in desc and row['ASV'] not in wildASVs:
            wildASVs.append(row['ASV'])
        if 'non_western_human' in desc and row['ASV'] not in nwhASVs:
            nwhASVs.append(row['ASV'])
        if 'western_human' in desc and row['ASV'] not in whASVs:
            whASVs.append(row['ASV'])
            
captiveASVs_df = allASVs[allASVs['ASV'].isin(captiveASVs)]

print(len(captiveASVs))
print(len(wildASVs))
print(len(nwhASVs))
print(len(whASVs))

captiveASVs_df.head()
```

```python
def select_seqs_fasta(fasta,select_fasta,list_of_ASVs):
    with open(fasta) as fa: 
        with open(select_fasta, 'w') as sel_fa:
            records = SeqIO.parse(fa, 'fasta')
            for record in records:
                if record.id in list_of_ASVs: 
                    SeqIO.write(record, sel_fa, 'fasta')
os.system('mkdir results/gyrb_all/Bt/select_fastas/')
AllASVs_fasta = 'results/gyrb_all/Bt/ASVs_filtered.fasta'

captiveASVs_fasta = 'results/gyrb_all/Bt/select_fastas/captive_apes_ASVs.fasta'
select_seqs_fasta(AllASVs_fasta,captiveASVs_fasta,captiveASVs)

wildASVs_fasta = 'results/gyrb_all/Bt/select_fastas/wild_apes_ASVs.fasta'
select_seqs_fasta(AllASVs_fasta,wildASVs_fasta,wildASVs)

nwhASVs_fasta = 'results/gyrb_all/Bt/select_fastas/nwh_apes_ASVs.fasta'
select_seqs_fasta(AllASVs_fasta,nwhASVs_fasta,nwhASVs)

whASVs_fasta = 'results/gyrb_all/Bt/select_fastas/wh_apes_ASVs.fasta'
select_seqs_fasta(AllASVs_fasta,whASVs_fasta,whASVs)
```

```python
%%bash 
cd results/gyrb_all/Bt/select_fastas

#make blastdb
makeblastdb -in wild_apes_ASVs.fasta -dbtype nucl
makeblastdb -in nwh_apes_ASVs.fasta -dbtype nucl
makeblastdb -in wh_apes_ASVs.fasta -dbtype nucl

#blast captive apes against wild apes, nwh, wh
blastn -query captive_apes_ASVs.fasta -db wild_apes_ASVs.fasta \
-outfmt "7 qseqid salltitles sseqid pident length qlen evalue" \
-out captive_vs_wild_blastout.txt \
-max_target_seqs 5
blastn -query captive_apes_ASVs.fasta -db nwh_apes_ASVs.fasta \
-outfmt "7 qseqid salltitles sseqid pident length qlen evalue" \
-out captive_vs_nwh_blastout.txt \
-max_target_seqs 5
blastn -query captive_apes_ASVs.fasta -db wh_apes_ASVs.fasta \
-outfmt "7 qseqid salltitles sseqid pident length qlen evalue" \
-out captive_vs_wh_blastout.txt \
-max_target_seqs 5
```

```python
def parse_blast(blast_res):
    blast_res = pd.read_csv(blast_res,sep='\t',comment='#',header=None)
    blast_res.columns = ['ASV','salltitles','sseqid','pident','length','qlen','evalue']
    blast_res = blast_res.groupby('ASV').head(1).reset_index(drop=True)
    return(blast_res)

captive_vs_wild = parse_blast('results/gyrb_all/Bt/select_fastas/captive_vs_wild_blastout.txt')   
captive_vs_wild = captive_vs_wild[['ASV','pident']]
captive_vs_wild.columns = ['ASV','wild_closest_pident']

captive_vs_nwh = parse_blast('results/gyrb_all/Bt/select_fastas/captive_vs_nwh_blastout.txt')   
captive_vs_nwh = captive_vs_nwh[['ASV','pident']]
captive_vs_nwh.columns = ['ASV','nwh_closest_pident']

captive_vs_wh = parse_blast('results/gyrb_all/Bt/select_fastas/captive_vs_wh_blastout.txt')   
captive_vs_wh = captive_vs_wh[['ASV','pident']]
captive_vs_wh.columns = ['ASV','wh_closest_pident']
```

```python
captiveASVs = captiveASVs_df.merge(captive_vs_wild,how='left',on='ASV')
captiveASVs = captiveASVs.merge(captive_vs_nwh,how='left',on='ASV')
captiveASVs = captiveASVs.merge(captive_vs_wh,how='left',on='ASV')
captiveASVs = captiveASVs.fillna('0')
captiveASVs.head()
```

```python
def ASV_designation(wild,nwh,wh,cutoff):
    if float(wild) >= cutoff and float(nwh) >= cutoff and float(wh) >= cutoff:
        return('panhominid')
    if float(wild) < cutoff and float(nwh) < cutoff and float(wh) < cutoff:
        return('unique_to_captive')
    if float(wild) >= cutoff and float(nwh) < cutoff and float(wh) < cutoff:
        return('wild')
    if float(wild) < cutoff and float(nwh) >= cutoff and float(wh) >= cutoff:
        return('human')
    return('other')
```

```python
ASV_designation(98,99,94,95)    
captiveASVs['ASV_designation'] = captiveASVs.apply(lambda row: ASV_designation(row['wild_closest_pident'],
                                                 row['nwh_closest_pident'],
                                                 row['wh_closest_pident'],
                                                 95),axis=1)
```
